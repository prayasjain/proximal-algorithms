```python
import matplotlib.pyplot as plt
import seaborn 
import cPickle as pickle
```

```python
with open('pga-score') as f:
    pga_score = pickle.load(f)
with open('pga-h') as f:
    pga_h = pickle.load(f)
```

```python
with open('fista-score') as f:
    fista_score = pickle.load(f)
with open('fista-h') as f:
    fista_h = pickle.load(f)
```

```python
with open('sgd-score') as f:
    sgd_score = pickle.load(f)
with open('sgd-h') as f:
    sgd_h = pickle.load(f)
```

```python
with open('sgd-acc-score') as f:
    sgd_acc_score = pickle.load(f)
with open('sgd-acc-h') as f:
    sgd_acc_h = pickle.load(f)
```

```python
with open('adam-score') as f:
    adam_score = pickle.load(f)
with open('adam-h') as f:
    adam_h = pickle.load(f)
```

```python
print len(pga_h['loss'])
```

Training Loss vs Epochs

```python
plt.figure(0)
plt.plot(list(range(1,41)),pga_h['loss'],'r')
plt.plot(list(range(1,41)),fista_h['loss'],'b')
plt.plot(list(range(1,41)),sgd_h['loss'],'g')
plt.plot(list(range(1,41)),sgd_acc_h['loss'],'c')
plt.plot(list(range(1,41)),adam_h['loss'],'m')
plt.xlabel('Epochs')
plt.ylabel('Optimization Function (Training)')
plt.legend(['PGA', 'FISTA', 'SGD', 'ACC. SGD','ADAM'], loc='upper right')
plt.show()
```

```python
plt.figure(0)
plt.plot(list(range(1,41)),pga_h['val_loss'],'r')
plt.plot(list(range(1,41)),fista_h['val_loss'],'b')
plt.plot(list(range(1,41)),sgd_h['val_loss'],'g')
plt.plot(list(range(1,41)),sgd_acc_h['val_loss'],'c')
plt.plot(list(range(1,41)),adam_h['val_loss'],'m')
plt.xlabel('Epochs')
plt.ylabel('Optimization Function (Testing)')
plt.legend(['PGA', 'FISTA', 'SGD', 'ACC. SGD','ADAM'], loc='upper right')
plt.show()
```

```python
plt.figure(0)
plt.plot(list(range(1,41)),pga_h['acc'],'r')
plt.plot(list(range(1,41)),fista_h['acc'],'b')
plt.plot(list(range(1,41)),sgd_h['acc'],'g')
plt.plot(list(range(1,41)),sgd_acc_h['acc'],'c')
plt.plot(list(range(1,41)),adam_h['acc'],'m')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (Training)')
plt.legend(['PGA', 'FISTA', 'SGD', 'ACC. SGD','ADAM'], loc='lower right')
plt.show()
```

```python
plt.figure(0)
plt.plot(list(range(1,41)),pga_h['val_acc'],'r')
plt.plot(list(range(1,41)),fista_h['val_acc'],'b')
plt.plot(list(range(1,41)),sgd_h['val_acc'],'g')
plt.plot(list(range(1,41)),sgd_acc_h['val_acc'],'c')
plt.plot(list(range(1,41)),adam_h['val_acc'],'m')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (Testing)')
plt.legend(['PGA', 'FISTA', 'SGD', 'ACC. SGD','ADAM'], loc='lower right')
plt.show()
```
